#### `sns.boxplot`

**Descripci贸n:** Crea un gr谩fico de caja que muestra la distribuci贸n de datos dividi茅ndolos en cuartiles.

**Componentes principales:**
- **Caja:** Representa el rango intercuart铆lico (IQR), que contiene el 50% de los datos.
- **L铆nea en la caja:** Representa la mediana.
- **"Bigotes":** Extienden hasta el valor m谩s peque帽o y m谩s grande dentro de 1.5 veces el IQR.
- **Puntos fuera de los bigotes:** Representan valores at铆picos (outliers).

**Par谩metros utilizados:**
- **`x='gender'` y `y='math_score'`:** Especifican las variables categ贸rica y num茅rica.
- **`hue='gender'`:** Agrega diferenciaci贸n por g茅nero usando colores.
- **`palette='Blues'`:** Define la paleta de colores.


```python
# Configurar el estilo del gr谩fico
sns.set(style="whitegrid")

# Crear un boxplot para las puntuaciones de matem谩ticas por g茅nero
plt.figure(figsize=(10, 6))  # Definir el tama帽o del gr谩fico

sns.boxplot(x='gender', y='math_score', data=df, palette="Blues", hue="gender")

# Agregar t铆tulo y etiquetas a los ejes
plt.title('Math Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Math Score')

# Mostrar el gr谩fico
plt.show()
```

El siguiente fragmento de c贸digo calcula una matriz de correlaci贸n para un conjunto de columnas relevantes del DataFrame y la visualiza como un mapa de calor (heatmap) utilizando **Seaborn** y **Matplotlib**.

###  `sns.heatmap`

**Descripci贸n:** Representa gr谩ficamente una matriz de correlaci贸n como un mapa de calor.

**Componentes principales:**
- **Celdas coloreadas:** Representan los coeficientes de correlaci贸n.
- **Anotaciones (`annot=True`):** Muestran los valores num茅ricos dentro de cada celda.
- **Paleta (`cmap='bwr'`):** Usa colores de azul a rojo para indicar relaciones negativas o positivas.
- **Par谩metros `vmin` y `vmax`:** Escalan el rango de correlaci贸n entre 0 y 1.
- **`linewidths=0.5`:** A帽ade bordes entre las celdas para mejorar la legibilidad.


```python
# Calcular la matriz de correlaci贸n para las columnas relevantes
correlation_matrix = df[['gender', 'lunch', 'test_preparation_course', 'math_score', 'reading_score', 'writing_score', 'total_score', 'average_score']].corr()

# Mostrar la matriz de correlaci贸n en consola
print(correlation_matrix)

# Configurar la figura de matplotlib para el mapa de calor
plt.figure(figsize=(10, 8))

# Crear el mapa de calor usando Seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='bwr', vmin=0, vmax=1, linewidths=0.5)

# Agregar t铆tulo y mostrar el gr谩fico
plt.title('Correlation Matrix of Student Performance Metrics')
plt.show()
```


Este fragmento de c贸digo utiliza la librer铆a **SciPy** para realizar una prueba T independiente. Se eval煤an las puntuaciones de matem谩ticas entre estudiantes masculinos y femeninos para determinar si existen diferencias estad铆sticamente significativas en sus medias.



```python
from scipy import stats

# Separar las puntuaciones de matem谩ticas por g茅nero
male_math_scores = df[df['gender'] == 1]['math_score']
female_math_scores = df[df['gender'] == 0]['math_score']

# Realizar la prueba T independiente
t_statistic, p_value = stats.ttest_ind(male_math_scores, female_math_scores)

# Mostrar los resultados
print(f"T-statistic: {t_statistic}, P-value: {p_value}")
```

Este c贸digo realiza un an谩lisis de varianza de una v铆a (**ANOVA**) utilizando la librer铆a **SciPy**. El objetivo es determinar si existen diferencias significativas en las puntuaciones promedio de los estudiantes seg煤n el nivel educativo de los padres.

- Filtra las puntuaciones promedio (average_score) para cada categor铆a de nivel educativo de los padres en el DataFrame.
- Pasa estas series como argumentos a la funci贸n f_oneway, que calcula:
    - F-statistic: Relaci贸n entre la variabilidad entre grupos y la variabilidad dentro de los grupos.
    - P-value: Probabilidad de que las diferencias observadas sean producto del azar.

```python
anova_result = stats.f_oneway(
    df[df['parental_level_of_education'] == "high school"]['average_score'],
    df[df['parental_level_of_education'] == "associate's degree"]['average_score'],
    df[df['parental_level_of_education'] == "bachelor's degree"]['average_score'],
    df[df['parental_level_of_education'] == "master's degree"]['average_score'],
    df[df['parental_level_of_education'] == "some college"]['average_score'],
    df[df['parental_level_of_education'] == "some high school"]['average_score']
)

print(f"ANOVA result: F-statistic = {anova_result.statistic}, P-value = {anova_result.pvalue}")
```
