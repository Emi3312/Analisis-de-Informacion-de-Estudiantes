### Uso de Seaborn y Matplotlib

El siguiente fragmento de código utiliza las librerías **Seaborn** y **Matplotlib** para crear una visualización de datos. En este caso, se genera un gráfico de caja (boxplot) para comparar las puntuaciones de matemáticas por género.

#### `sns.boxplot`

**Descripción:** Crea un gráfico de caja que muestra la distribución de datos dividiéndolos en cuartiles.

**Componentes principales:**
- **Caja:** Representa el rango intercuartílico (IQR), que contiene el 50% de los datos.
- **Línea en la caja:** Representa la mediana.
- **"Bigotes":** Extienden hasta el valor más pequeño y más grande dentro de 1.5 veces el IQR.
- **Puntos fuera de los bigotes:** Representan valores atípicos (outliers).

**Parámetros utilizados:**
- **`x='gender'` y `y='math_score'`:** Especifican las variables categórica y numérica.
- **`hue='gender'`:** Agrega diferenciación por género usando colores.
- **`palette='Blues'`:** Define la paleta de colores.


```python
# Configurar el estilo del gráfico
sns.set(style="whitegrid")

# Crear un boxplot para las puntuaciones de matemáticas por género
plt.figure(figsize=(10, 6))  # Definir el tamaño del gráfico

sns.boxplot(x='gender', y='math_score', data=df, palette="Blues", hue="gender")

# Agregar título y etiquetas a los ejes
plt.title('Math Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Math Score')

# Mostrar el gráfico
plt.show()
```

El siguiente fragmento de código utiliza **Seaborn** y **Matplotlib** para generar gráficos de caja (boxplots) que comparan las puntuaciones de lectura y escritura por género.


```python
# Boxplot para las puntuaciones de lectura por género
plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='reading_score', data=df, palette="Greens", hue="gender")
plt.title('Reading Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Reading Score')
plt.show()

# Boxplot para las puntuaciones de escritura por género
plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='writing_score', data=df, palette="Oranges", hue="gender")
plt.title('Writing Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Writing Score')
plt.show()
```

El siguiente fragmento de código calcula la puntuación promedio de tres materias (matemáticas, lectura y escritura) para cada nivel de educación de los padres y visualiza los resultados utilizando un gráfico de barras.


```python
# Calcular la puntuación promedio en matemáticas, lectura y escritura para cada nivel de educación parental
df['average_score'] = df[['math_score', 'reading_score', 'writing_score']].mean(axis=1)

# Crear un gráfico de barras de las puntuaciones promedio por nivel de educación parental
plt.figure(figsize=(12, 6))
sns.barplot(x='parental_level_of_education', y='average_score', data=df, palette="coolwarm")
plt.title('Average Scores by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.show()
```

El siguiente fragmento de código genera un gráfico de caja (boxplot) para comparar las puntuaciones promedio de los estudiantes según si completaron o no un curso de preparación para los exámenes.


```python
# Boxplot para la puntuación promedio según la finalización del curso de preparación para exámenes (1 = completado, 0 = no completado)
plt.figure(figsize=(10, 6))
sns.boxplot(x='test_preparation_course', y='average_score', data=df, palette="Set2", hue="test_preparation_course")
plt.title('Average Scores by Test Preparation Course Completion')
plt.xlabel('Test Preparation Course (1 = Completed, 0 = Not Completed)')
plt.ylabel('Average Score')
plt.show()
```

El siguiente fragmento de código calcula una matriz de correlación para un conjunto de columnas relevantes del DataFrame y la visualiza como un mapa de calor (heatmap) utilizando **Seaborn** y **Matplotlib**.

###  `sns.heatmap`

**Descripción:** Representa gráficamente una matriz de correlación como un mapa de calor.

**Componentes principales:**
- **Celdas coloreadas:** Representan los coeficientes de correlación.
- **Anotaciones (`annot=True`):** Muestran los valores numéricos dentro de cada celda.
- **Paleta (`cmap='bwr'`):** Usa colores de azul a rojo para indicar relaciones negativas o positivas.
- **Parámetros `vmin` y `vmax`:** Escalan el rango de correlación entre 0 y 1.
- **`linewidths=0.5`:** Añade bordes entre las celdas para mejorar la legibilidad.


```python
# Calcular la matriz de correlación para las columnas relevantes
correlation_matrix = df[['gender', 'lunch', 'test_preparation_course', 'math_score', 'reading_score', 'writing_score', 'total_score', 'average_score']].corr()

# Mostrar la matriz de correlación en consola
print(correlation_matrix)

# Configurar la figura de matplotlib para el mapa de calor
plt.figure(figsize=(10, 8))

# Crear el mapa de calor usando Seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='bwr', vmin=0, vmax=1, linewidths=0.5)

# Agregar título y mostrar el gráfico
plt.title('Correlation Matrix of Student Performance Metrics')
plt.show()
```


Este fragmento de código utiliza la librería **SciPy** para realizar una prueba T independiente. Se evalúan las puntuaciones de matemáticas entre estudiantes masculinos y femeninos para determinar si existen diferencias estadísticamente significativas en sus medias.



```python
from scipy import stats

# Separar las puntuaciones de matemáticas por género
male_math_scores = df[df['gender'] == 1]['math_score']
female_math_scores = df[df['gender'] == 0]['math_score']

# Realizar la prueba T independiente
t_statistic, p_value = stats.ttest_ind(male_math_scores, female_math_scores)

# Mostrar los resultados
print(f"T-statistic: {t_statistic}, P-value: {p_value}")
```

Este código realiza un análisis de varianza de una vía (**ANOVA**) utilizando la librería **SciPy**. El objetivo es determinar si existen diferencias significativas en las puntuaciones promedio de los estudiantes según el nivel educativo de los padres.

- Filtra las puntuaciones promedio (average_score) para cada categoría de nivel educativo de los padres en el DataFrame.
- Pasa estas series como argumentos a la función f_oneway, que calcula:
    - F-statistic: Relación entre la variabilidad entre grupos y la variabilidad dentro de los grupos.
    - P-value: Probabilidad de que las diferencias observadas sean producto del azar.

```python
anova_result = stats.f_oneway(
    df[df['parental_level_of_education'] == "high school"]['average_score'],
    df[df['parental_level_of_education'] == "associate's degree"]['average_score'],
    df[df['parental_level_of_education'] == "bachelor's degree"]['average_score'],
    df[df['parental_level_of_education'] == "master's degree"]['average_score'],
    df[df['parental_level_of_education'] == "some college"]['average_score'],
    df[df['parental_level_of_education'] == "some high school"]['average_score']
)

print(f"ANOVA result: F-statistic = {anova_result.statistic}, P-value = {anova_result.pvalue}")
```
